```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
```

```python
df.head(10)
```

```python
sns.distplot(df['petal_length'])
```

```python
sns.distplot(df['petal_width'],kde=False)
```

```python
sns.distplot(df['sepal_width'])
```

```python
sns.distplot(df['sepal_length'],hist= False)
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data= df)
```

```python
sns.pairplot(df,hue='species')
```

```python
sns.jointplot(x='petal_length',y='petal_width',data= df)
```

```python
sns.stripplot(x='species',y='petal_length',data=df,jitter=True)
```

```python
sns.barplot(x='species',y='petal_length',data=df)
```

```python
sns.boxplot(data= df)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot =True)
```
